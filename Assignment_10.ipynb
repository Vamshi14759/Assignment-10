```python
## Objective : The objective of this assignment is to apply Association Rule Mining using the Apriori algorithm on the Online Retail dataset to discover meaningful product purchase patterns.
```

```python
## Dataset    Online Retail Dataset (UCI style transactional data).
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Load the Excel file. Based on previous output, it's being read as a single column
# where the first cell is the column header, and subsequent cells are transaction strings.
df_raw = pd.read_excel('Online retail.xlsx', header=None)

# Assuming the actual transactions are in the first column (index 0).
# Each entry is a string of comma-separated items.

# We need to extract the first row as a transaction, and then the rest of the column as transactions.
# The column name from the previous `df` output was 'shrimp,almonds,...'
# and the first data row was 'burgers,meatballs,eggs'.

# If the first row of the Excel file (now df_raw.iloc[0,0]) is meant to be a transaction itself,
# and not a header, we can process all rows.

# Extract all transaction strings from the first column
all_transaction_strings = df_raw.iloc[:, 0].astype(str).tolist()

# Process each transaction string into a list of cleaned items
transactions = []
for trans_str in all_transaction_strings:
    # Split by comma, strip whitespace, and filter out empty strings
    items = [item.strip() for item in trans_str.split(',') if item.strip()]
    if items:  # Add only non-empty transaction lists
        transactions.append(items)

# The variable 'transactions' is now a list of lists, suitable for TransactionEncoder.
# The original cleaning steps (`dropna` on `InvoiceNo`, filtering cancelled transactions,
# and cleaning `Description`) are not applicable to this data format and have been removed.
```

```python
# Use the 'transactions' list prepared in the previous cell directly.
# The original code for grouping by 'InvoiceNo' and 'Description' is not needed
# because the 'transactions' variable is already a list of lists, where each inner
# list represents a complete transaction (basket of items).
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
basket = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
rules.sort_values(by='lift', ascending=False).head()
```
